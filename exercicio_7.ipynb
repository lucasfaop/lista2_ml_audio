```
def bark_1990(freq):
  resp = (26.81*freq)/(1960+freq) - 0.53
  return resp
```

```
import math
def mel_htk(freq):
  resp = 2595*math.log10(1 + freq/700)
  return resp
```

```
print("Item a")
print("Bark: " + str(bark_1990(100)) +" barks")
print("Mel: " + str(mel_htk(100)) +" mels")
```
