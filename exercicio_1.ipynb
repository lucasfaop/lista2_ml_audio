```
import librosa
import librosa.display
# matplotlib for displaying the output
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Item a

```
f = 2
t = np.arange(0,5.1, 0.01)
x = np.sin(2 * np.pi * f * t)
```

```
fig, ax = plt.subplots()
ax.plot(t,x)
ax.set_xlabel('Tempo [s]')
ax.set_ylabel('Amplitude')
```

# Item B

```
f1 = 1
f2 = 2
N =500 
t = np.arange(0,5.1, 0.01)
t1 = t[:N//2]
t2 = t[N//2:]
x1 = 1.0 * np.sin(2 * np.pi * f1 * t1)
x2 = 1.0 * np.sin(2 * np.pi * f2 * t2- np.pi)
x = np.concatenate((x1, x2))
```

```
plt.plot(t, x)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
```

# Item C

```
f1 = 2
f2 = 1
N =500 
t = np.arange(0,5.1, 0.01)
t1 = t[:2*N//5]
t2 = t[3*N//5:]
x1 = 1.0 * np.sin(2 * np.pi * f1 * t1)
x2 = 1.0 * np.sin(2 * np.pi * f2 * t2)
x3 = [0 for i in range(100)]
x = np.concatenate((x1, x3, x2))
```

```
plt.plot(t, x)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
```

# Item d

```
from scipy.signal import chirp
f = 1
t = np.arange(0,5.1, 0.001)
x =  chirp(t, f0=2, f1=25, t1=5, method='linear')
```
