```
import numpy as np
```

```
def short_time_energy(vetor):
  soma = 0
  for v in vetor:
    soma += v**2
  return soma

def short_time_zero_crossing(vetor):
  cortes = 0
  for i in range(1, len(vetor)):
    if vetor[i]>0 and vetor[i-1]<0:
      cortes+= 1
    elif vetor[i]<0 and vetor[i-1]>0:
      cortes+= 1
  return cortes
```

```
def window(a, w, o, copy = True):
    sh = (a.size - w + 1, w)
    st = a.strides * 2
    view = np.lib.stride_tricks.as_strided(a, strides = st, shape = sh)[0::o]
    if copy:
        return view.copy()
    else:
        return view
```

```
def calc_short_times(vetor, w, o):
  win = window(vetor, w, o, copy = True)
  senergy = []
  szerocross = []
  for w in win:
    senergy.append(short_time_energy(w))
    szerocross.append(short_time_zero_crossing(w))
  return np.array(senergy), np.array(szerocross)
```
