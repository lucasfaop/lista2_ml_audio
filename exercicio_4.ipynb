```
def calc_short_times(vetor, w, o):
  return librosa.feature.rms(vetor,frame_length = w, hop_length = o)[0], librosa.feature.zero_crossing_rate(vetor,frame_length = w, hop_length = o)[0]
```

```
x.shape
```

```
energia, cruza_zero = calc_short_times(x, 128, 64)
```

```
librosa.display.waveshow(x,sr)
```

```
t = [i for i in range(len(energia))]
import matplotlib.pyplot as plt


plt.figure(figsize=(20, 7))
plt.subplot(3, 1, 1)
librosa.display.waveshow(x,sr)
plt.title('Short-Time Energy')

plt.subplot(3, 1, 2)
plt.plot(t, energia, 'r')
plt.title('Short-Time Energy')

plt.subplot(3, 1, 3)
plt.plot(t, cruza_zero, 'g')
plt.title('Short-Time Zero-Crossing')
plt.tight_layout()
plt.show()
```
